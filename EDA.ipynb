```python
import pandas as pd
data=pd.read_csv("supermarket_data.csv")
data
```

```python
# decribe the data
data.describe()
```

```python
# fill empty item price
data["itemprice"]=data["itemprice"].fillna(55)
data.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# count plot to show didtribution of data
# 1 which gender has more transactions
# 2 whats your observation what do you recommend


sns.countplot(x="gender",data=data,palette="dark")
```

```python
# count plot for products
# which product categories appears most/least
# customer buying behaviour
# Recomendation
sns.countplot(x="product",data=data,palette="dark")
plt.xticks(rotation=45)
```

```python
# countplt visualizing distribution of customers by gender accross diffrent cities
# 1. which gender is highest
# 2. which city has least and the most customers beased on gender
# 3. wht do you observe about gender
# 4. what is your recomendation

sns.countplot(x="gender",data=data,palette="dark",hue="city")
```

```python
# pie chart
# distribution of customers by paymentmethod
# 1 . the most and least used payment method
# 2. Percentages of each payment method 
# 3.which payment methods are similar
# 4.what may infulence the prefernce of most used method
# 5 .whst your recommendation
data["payment"].value_counts().plot.pie(autopct='%1.1f%%')
```
